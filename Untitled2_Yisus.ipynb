```python
import sympy as sp

t = sp.symbols('t')
y = sp.Function('y')(t)

# Resolver la nueva ecuaci√≥n: dy/dt + 4y = e^(-4t), y(0)=2
sol = sp.dsolve(
    sp.Eq(y.diff(t) + 4*y, sp.exp(-4*t)),
    ics={y.subs(t, 0): 2}
)

sp.init_printing(use_latex='mathjax')
print("Soluci√≥n:")
display(sol)
```

```python
import sympy as sp

t = sp.symbols('t')
y = sp.Function('y')(t)

# Resolver directamente con condiciones iniciales
sol = sp.dsolve(
    sp.Eq(y.diff(t, t) + 5*y.diff(t) + 4*y, 0),
    ics={y.subs(t, 0): 1, y.diff(t).subs(t, 0): 0}
)

sp.init_printing(use_latex='mathjax')
display(sol)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal

def analisis_circuito_protector():

    # Par√°metros del circuito
    R = 10.0           # Œ©
    L = 0.1            # H
    C = 100e-6         # F = 0.0001 F
    Vin = 5.0          # V

    print("\n" + "MODELADO MATEM√ÅTICO CON LAPLACE")
    print("-" * 50)

    # C√°lculos de coeficientes
    LC = L * C
    RC = R * C
    # =========================================================================
    #   Modelo Matematico
    # =========================================================================

    print("Ecuaci√≥n diferencial en el dominio del tiempo:")
    print(f"LC¬∑d¬≤v_C/dt¬≤ + RC¬∑dv_C/dt + v_C = V_in(t)")
    print(f"({L})({C})¬∑d¬≤v_C/dt¬≤ + ({R})({C})¬∑dv_C/dt + v_C = {Vin}u(t)")
    print(f"{LC:.6f}¬∑d¬≤v_C/dt¬≤ + {RC:.4f}¬∑dv_C/dt + v_C = {Vin}u(t)")

    print("\nAplicando Transformada de Laplace (Solucion en dominio s):")
    print(f"V_C(s) = {Vin} / [s({LC:.6f}s¬≤ + {RC:.4f}s + 1)]")

    H_s = f"1/({LC:.6f}s¬≤ + {RC:.4f}s + 1)"
    print(f"Funci√≥n de transferencia (relacion entrada/salida): H(s) = {H_s}")

    # =========================================================================
    #  Respuesta Temporal
    # =========================================================================

    print("\n" + "Respuesta temporal v_C(t)")
    print("-" * 50)

    # Par√°metros del sistema
    w0 = 1 / np.sqrt(L * C)
    zeta = (R / 2) * np.sqrt(C / L)
    wd = w0 * np.sqrt(1 - zeta**2)
    alpha = zeta * w0

    print("Par√°metros calculados:")
    print(f"‚Ä¢ Frecuencia natural: œâ‚ÇÄ = {w0:.2f} rad/s")
    print(f"‚Ä¢ Factor amortiguamiento: Œ∂ = {zeta:.3f}")
    print(f"‚Ä¢ Tipo: {'SUBAMORTIGUADO' if zeta < 1 else 'SOBREAMORTIGUADO'}")

    # Soluci√≥n temporal
    if zeta < 1:
        print("\nSoluci√≥n temporal v_C(t):")
        print(f"v_C(t) = {Vin} - e^(-{alpha:.1f}t)[{Vin}¬∑cos({wd:.1f}t) + {Vin*R*C*w0**2/wd:.1f}¬∑sin({wd:.1f}t)] V")

    # Simulaci√≥n num√©rica
    t = np.linspace(0, 0.1, 1000)
    num = [1]
    den = [LC, RC, 1]  # Usando los coeficientes CORRECTOS
    _, vc = signal.step((num, den), T=t)
    vc = Vin * vc

    vc_max = np.max(vc)
    vc_final = vc[-1]
    sobreimpulso = ((vc_max - vc_final) / vc_final) * 100

    print(f"\nResultados num√©ricos:")
    print(f"‚Ä¢ Voltaje m√°ximo: {vc_max:.2f} V")
    print(f"‚Ä¢ Voltaje final: {vc_final:.2f} V")
    print(f"‚Ä¢ Sobreimpulso: {sobreimpulso:.1f}%")

    # =========================================================================
    #  Gr√°ficas y an√°lisis
    # =========================================================================
    print("\n" + "Graficas y analisis")
    print("-" * 50)

    # Configurar gr√°ficas
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 12))

    # 3a) Respuesta temporal
    ax1.plot(t, vc, 'orange', linewidth=2, label='Voltaje v_C(t)')
    ax1.axhline(y=Vin, color='green', linestyle='--', alpha=0.7, label='5 V (estado estable)')
    ax1.axhline(y=vc_max, color='blue', linestyle='--', alpha=0.7, label=f'M√°ximo: {vc_max:.2f} V')
    ax1.set_xlabel('Tiempo (s)')
    ax1.set_ylabel('Voltaje v_C(t) (V)')
    ax1.set_title('a) Respuesta Temporal - Evoluci√≥n de v_C(t)')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # 3b) Respuesta en frecuencia - Magnitud
    f = np.logspace(0, 4, 1000)  # 1 Hz a 10 kHz
    w = 2 * np.pi * f
    H = 1 / (LC*(1j*w)**2 + RC*(1j*w) + 1)  # Usando coeficientes CORRECTOS

    magnitude_dB = 20 * np.log10(np.abs(H))

    ax2.semilogx(f, magnitude_dB, 'purple', linewidth=2)
    ax2.axhline(y=-5, color='black', linestyle='--', alpha=0.7, label='-3 dB')

    # Encontrar frecuencia de corte
    cutoff_idx = np.argmin(np.abs(magnitude_dB + 3))
    f_cutoff = f[cutoff_idx]
    ax2.axvline(x=f_cutoff, color='orange', linestyle='--', alpha=0.7,
                label=f'Frecuencia de Corte: {f_cutoff:.1f} Hz')

    ax2.set_xlabel('Frecuencia (Hz)')
    ax2.set_ylabel('Magnitud (dB)')
    ax2.set_title('b) Respuesta en Frecuencia - Magnitud')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    ax2.set_ylim([-60, 5])

    # 3c) Respuesta en frecuencia - Fase
    phase_deg = np.angle(H, deg=True)
    ax3.semilogx(f, phase_deg, 'teal', linewidth=2)
    ax3.set_xlabel('Frecuencia (Hz)')
    ax3.set_ylabel('Fase (grados)')
    ax3.set_title('c) Respuesta en Frecuencia - Fase')
    ax3.grid(True, alpha=0.3)
    ax3.set_ylim([-180, 10])

    plt.tight_layout()
    plt.show()

    print("Gr√°ficas generadas:")
    print("‚Ä¢ a) v_C(t) vs tiempo - Transitorio y estado estable")
    print("‚Ä¢ b) Magnitud vs frecuencia - Comportamiento como filtro")
    print("‚Ä¢ c) Fase vs frecuencia - Desfase del circuito")
    print(f"‚Ä¢ Frecuencia de corte: {f_cutoff:.1f} Hz")

    # =========================================================================
    # OBJETIVO 4: Interpretaci√≥n f√≠sica
    # =========================================================================
    print("\n" + "INTERPRETACI√ìN F√çSICA")
    print("-" * 50)

    # An√°lisis de protecci√≥n
    atenuacion_10kHz = magnitude_dB[-1]

    print("a) ¬øProtege el router de picos o ruido?")
    print(f"   ‚úì S√ç - Aten√∫a {abs(atenuacion_10kHz):.1f} dB a 10 kHz")
    print(f"   ‚úì Comportamiento: FILTRO PASA-BAJOS")
    print(f"   ‚úì Frecuencias > {f_cutoff:.0f} Hz son atenuadas")

    print("\nb) ¬øExiste sobreimpulso (voltaje > 5V)?")
    print(f"   ‚úì S√ç - Sobreimpulso de {sobreimpulso:.1f}%")
    print(f"   ‚úì Voltaje m√°ximo: {vc_max:.2f} V (riesgo potencial)")

    print("\nc) Efecto de cambiar par√°metros:")

    # An√°lisis de sensibilidad
    print("   ‚Ä¢ Reducir R a 1Œ©:")
    zeta_1 = (1 / 2) * np.sqrt(C / L)
    if zeta_1 < 1:
        Mp_1 = np.exp(-np.pi * zeta_1 / np.sqrt(1 - zeta_1**2)) * 100
        print(f"     ‚Üí Œ∂ = {zeta_1:.3f}, Mp = {Mp_1:.1f}% (MUY PELIGROSO)")

    print("   ‚Ä¢ Aumentar C a 470ŒºF:")
    w0_470 = 1 / np.sqrt(L * 470e-6)
    zeta_470 = (R / 2) * np.sqrt(470e-6 / L)
    if zeta_470 < 1:
        Mp_470 = np.exp(-np.pi * zeta_470 / np.sqrt(1 - zeta_470**2)) * 100
        print(f"     ‚Üí f‚ÇÄ = {w0_470/(2*np.pi):.1f} Hz, Mp = {Mp_470:.1f}% (m√°s protecci√≥n)")


if __name__ == "__main__":
    analisis_circuito_protector()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal

# Par√°metros del circuito RLC original
R = 10.0
L = 0.1
C = 100e-6
Vin = 5.0

# Crear figura con dos subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# =============================================================================
# GR√ÅFICA (a): DOMINIO DEL TIEMPO
# =============================================================================
ax1.set_title('(a) Dominio de tiempo - Tensi√≥n en el Capacitor', fontsize=14, fontweight='bold')

# Calcular respuesta temporal
t = np.linspace(0, 0.1, 1000)  # 100 ms
num = [1]
den = [L*C, R*C, 1]
_, vc_response = signal.step((num, den), T=t)
vc_response = Vin * vc_response

# Graficar respuesta temporal
ax1.plot(t * 1000, vc_response, 'orange', linewidth=2.5, label='$v_C(t)$ - Tensi√≥n en el router')  # Cambiado a naranja
ax1.axhline(y=Vin, color='blue', linestyle='--', linewidth=1.5, alpha=0.7, label='5 V (Estado estable)')  # Cambiado a azul

# Marcar caracter√≠sticas importantes
vc_max = np.max(vc_response)
t_max = t[np.argmax(vc_response)] * 1000
ax1.axhline(y=vc_max, color='purple', linestyle='--', linewidth=1, alpha=0.6, label=f'M√°ximo: {vc_max:.2f} V')  # Cambiado a p√∫rpura

ax1.set_xlabel('Tiempo (ms)', fontsize=12)
ax1.set_ylabel('Tensi√≥n $v_C(t)$ (V)', fontsize=12)
ax1.legend(loc='lower right', fontsize=10)
ax1.grid(True, alpha=0.3)
ax1.set_xlim(0, 100)

# Anotaciones
ax1.annotate('Subida suave con oscilaci√≥n', xy=(20, 6), xytext=(40, 7.5),
            arrowprops=dict(arrowstyle='->', color='red'), fontsize=10)
ax1.annotate('Sobreimpulso del 60%', xy=(t_max, vc_max), xytext=(50, 8),
            arrowprops=dict(arrowstyle='->', color='purple'), fontsize=10)

# =============================================================================
# GR√ÅFICA (b): DOMINIO DE LA FRECUENCIA
# =============================================================================
ax2.set_title('(b) Respuesta en Frecuencia - Comportamiento como Filtro', fontsize=14, fontweight='bold')

# Calcular respuesta en frecuencia
f = np.logspace(0, 5, 1000)  # 1 Hz to 100 kHz
w = 2 * np.pi * f
H = 1 / (L*C*(1j*w)**2 + R*C*(1j*w) + 1)
magnitude_dB = 20 * np.log10(np.abs(H))

# Graficar magnitud en dB
ax2.semilogx(f, magnitude_dB, 'teal', linewidth=2.5, label='Magnitud (dB)')  # Cambiado a teal

# L√≠neas de referencia
ax2.axhline(y=-3, color='gray', linestyle='--', linewidth=1, alpha=0.7, label='-3 dB')  # Cambiado a gris
ax2.axvline(x=50, color='magenta', linestyle='--', linewidth=1, alpha=0.7, label='Frecuencia de corte ‚âà 50 Hz')  # Cambiado a magenta

# Zonas de frecuencia
ax2.axvspan(1, 50, alpha=0.2, color='lightgreen', label='Frecuencias pasan')  # Cambiado a verde claro
ax2.axvspan(50, 1e5, alpha=0.2, color='lightcoral', label='Frecuencias atenuadas')  # Cambiado a coral claro

ax2.set_xlabel('Frecuencia (Hz)', fontsize=12)
ax2.set_ylabel('Magnitud (dB)', fontsize=12)
ax2.legend(loc='upper right', fontsize=10)
ax2.grid(True, alpha=0.3)
ax2.set_ylim(-60, 5)
ax2.set_xlim(1, 1e5)

# Anotaciones
ax2.annotate('Filtro Pasa-Bajos', xy=(100, -10), xytext=(500, 0),
            arrowprops=dict(arrowstyle='->', color='blue'), fontsize=12, fontweight='bold')
ax2.annotate('Aten√∫a ruido\nde alta frecuencia', xy=(1e4, -30), xytext=(2e3, -15),
            arrowprops=dict(arrowstyle='->', color='red'), fontsize=10)

plt.tight_layout()
plt.show()

# =============================================================================
# INFORMACI√ìN ADICIONAL PARA INTERPRETACI√ìN
# =============================================================================
print("=" * 70)
print("INTERPRETACI√ìN DE LAS GR√ÅFICAS")
print("=" * 70)
print("\nGR√ÅFICA (a) - DOMINIO DEL TIEMPO:")
print("‚Ä¢ La tensi√≥n en el capacitor sube suavemente de 0V a 5V")
print("‚Ä¢ Presenta una peque√±a oscilaci√≥n con sobreimpulso del 60%")
print("‚Ä¢ El voltaje m√°ximo alcanza ‚âà 8V (riesgo potencial)")
print("‚Ä¢ Despu√©s de ‚âà 80ms se estabiliza en 5V")

print("\nGR√ÅFICA (b) - DOMINIO DE LA FRECUENCIA:")
print("‚Ä¢ Comportamiento t√≠pico de FILTRO PASA-BAJOS")
print("‚Ä¢ Frecuencias bajas (< 50 Hz) pasan casi sin atenuaci√≥n")
print("‚Ä¢ Frecuencias altas (> 50 Hz) son atenuadas progresivamente")
print("‚Ä¢ Aten√∫a ruido el√©ctrico de alta frecuencia protegiendo el router")
print("‚Ä¢ Pendiente de -40 dB/d√©cada t√≠pica de filtros de segundo orden")

print(f"\nPAR√ÅMETROS CLAVE:")
print(f"‚Ä¢ Frecuencia de corte: ‚âà 50 Hz")
print(f"‚Ä¢ Atenuaci√≥n a 10 kHz: {magnitude_dB[f > 10000][0]:.1f} dB")
print(f"‚Ä¢ Sobreimpulso temporal: {((vc_max - Vin)/Vin*100):.1f}%")
```
